Implement function triangleArea(a,b,c) that takes as input the lengths of the 3 sides of a triangle and returns the area of the triangle. By Heron's formula, the area of a triangle with side lengths a, b, and c is

s(s - a)(s -b)(s -c)

, where s = (a+b+c)/2.

triangleArea(2,2,2)

1.7320508075688772

```sqlite3
using System;

namespace Dataprogramming_Question3
{
    class Program
    {
        static void Main(string[] args)
        {
            double s, x, y, z, area;

            Console.WriteLine("Enter the values of sides of triangle x, y, z \n");
            x = Convert.ToDouble(Console.ReadLine());
            y = Convert.ToDouble(Console.ReadLine());
            z = Convert.ToDouble(Console.ReadLine());

            s = (x + y + z) / 2;

            area = Math.Sqrt(s * (s - x) * (s - y) * (s - z));

            Console.WriteLine("Area of triangle is " + area);
            Console.ReadLine();

        }
    }
}
```

Question 4 C#
Write a program in C# Sharp to separate odd and even integers in separate arrays.
Go to the editor
Test Data :
Input the number of elements to be stored in the array :5
Input 5 elements in the array :
element - 0 : 25
element - 1 : 47
element - 2 : 42
element - 3 : 56
element - 4 : 32 

```sqlite3
using System;

namespace Problem_set1
{
    class Program
    {
        static void Main(string[] args)
        {
            Console.WriteLine("Hello World!");

            int[] collector = new int[10];
            int[] evenArray = new int[10];
            int[] oddArray = new int[10];
            int x, y = 0, z = 0, i;


            Console.Write("Input the number of elements to be stored in the array :");
            i = Convert.ToInt32(Console.ReadLine());

            Console.Write("Input {0} elements in the array :\n", i);
            for (x = 0; x < i; x++)
            {
                Console.Write("element - {0} : ", x);
                collector[x] = Convert.ToInt32(Console.ReadLine());
            }

            for (x = 0; x < i; x++)
            {
                if (collector[x] % 2 == 0)
                {
                    evenArray[y] = collector[x];
                    y++;
                }
                else
                {
                    oddArray[z] = collector[x];
                    z++;
                }
            }

            Console.Write("\nThe Even elements are : \n");
            for (x = 0; x < y; x++)
            {
                Console.Write("{0} ", evenArray[x]);
            }

            Console.Write("\nThe Odd elements are :\n");
            for (x = 0; x < z; x++)
            {
                Console.Write("{0} ", oddArray[x]);
            }
            Console.ReadLine();

        }
    }
}
```
